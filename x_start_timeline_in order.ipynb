```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob

# set the folder path and suffix
folder_path = "C:/Users/uqyrong/Desktop/test"
suffix = "*_peak_baseline.xlsx"
excel_files = glob.glob(folder_path + "/" + suffix)

for file in excel_files:
    df = pd.read_excel(file, sheet_name="peak", usecols=[0, 1])
    names= df.iloc[:, 0]
    timepoints = df.iloc[:, 1]
# print(names, timepoints)

levels = np.tile([-0.9, -0.7, -0.5, -0.3, 0.3, 0.5, 0.7, 0.9],
                 int(np.ceil(len(timepoints)/8)))[:len(timepoints)]

# Create figure and plot a stem plot with the date
fig, ax = plt.subplots(figsize=(20, 6))
ax.set(title="triple mutant_all neurons order")

ax.vlines(timepoints, 0, levels, color="tab:red")  # The vertical stems.
ax.plot(timepoints, np.zeros_like(timepoints), "-o",
        color="k", markerfacecolor="w")  # Baseline and markers on it.

# annotate lines
for t, l, n in zip(timepoints, levels, names):
    ax.annotate(n, xy=(t, l),
                xytext=(-3, np.sign(l)*3), textcoords="offset points",
                horizontalalignment="right",
                verticalalignment="bottom" if l > 0 else "top", fontsize=10)
plt.savefig('figure.jpg', dpi=300, bbox_inches="tight")
plt.show()
```
